```python
import qiskit
import qiskit_aer
from qiskit.quantum_info import Statevector
from qiskit.circuit.library import MCXGate
from qiskit_aer import QasmSimulator
```

```python
#Choose the number of qubits you want to test on, including the auxiliary qubit

n= 5

qr = qiskit.QuantumRegister(n-1, name= "qubit")
aux = qiskit.QuantumRegister(1,name="aux")
cr = qiskit.ClassicalRegister(n-1,name="bit")
qc = qiskit.QuantumCircuit(qr,aux,cr)
display(qc.draw("mpl"))
```

```python
#Acting with X-Gate on the auxiliary qubit.
qc.x(aux)

qc.barrier()
display(qc.draw("mpl"))

#Reversing the circuit just to make it read qubits from top-to-bottom.
reversed_qc = qc.reverse_bits()
psi1 = Statevector.from_instruction(reversed_qc)
display("Psi1", psi1.draw('latex'))
```

```python
#Applying Hadamard Gate on all qubits
for i in range(len(qr)+1):
    qc.h(i)
    
qc.barrier()
display(qc.draw("mpl"))

reversed_qc = qc.reverse_bits()
psi2 = Statevector.from_instruction(reversed_qc)
display("Psi2", psi2.draw('latex'))
```

Since the Oracle are blackboxes, we will not be able to find an exact unitary operator that is required for the Deutsh's Jozsa Algorithm. Instead we will be able to mimic the logic of the gate based on the type of the oracle.

We will build an oracle for each possible function type, as guarranteed by the algorithm, that the function will be either "constant" or "balanced", so we have then 4 cases that we need to take into consideration:

I- Oracle function is constant:
    I.1 function return -> f(i) = 0
    I.2 function return -> f(i) = 1

II- Oracle function is balanced:
    II.1 function return -> f(0) = 0 & f(1) = 1
    II.2 function return -> f(0) = 1 & f(1) = 0


For Each of these cases, the "Oracle" will have a different impact on the circuit. And in order to achieve the same results we have to use different Unitary operators for each case, note that in order to find out what gates lead to the same results, i would recommend solving "Deutsh's Jozsa" circuit manually, anyway here are the different unitary operators for each case:

I.1 function return -> f(i) = 0 : 
    In this case, nothing will change so we will only apply Identity gate on the auxiliary qubit.
    
I.2 function return -> f(i) = 1 : 
    In this case, we will have to flip the auxiliary qubit, so we will apply X-Gate on the auxiliary qubit.
    
II.1 function return -> f(0) = 0 & f(1) = 1 : 
    In this case, we will have to flip the auxiliary qubit only if i=1 so that f(1) = 1, and this can be modeled by a CX-Gate, where the any of the top qubits is the control qubit and the target qubit is the auxiliary qubit. 
    
II.2 function return -> f(0) = 1 & f(1) = 0 : 
    In this case, we will have to flip the auxiliary qubit only if i=0 so that f(0) = 1, and this can be modeled by an inverted CX-Gate, where the any of the top qubits is the control qubit and the target qubit is the auxiliary qubit. 



```python
import random

# Define the different oracles types

oracles = ("constant","balanced")
constant_oracles = ("constant returning zeros f(i)=0","constant returning ones, f(i)=1")
balanced_oracles = ("Balanced function: f(0) = 0 & f(1) = 1","Balanced function: f(1) = 0 & f(0)=1")

# Pick a random value for the oracle, so we can test for different cases
Random_oracle = random.choice(oracles)

if Random_oracle == "constant":
    Random_oracle = random.choice(constant_oracles)
else:
    Random_oracle = random.choice(balanced_oracles)

print(Random_oracle)


if Random_oracle == "constant returning zeros f(i)=0":
    qc.id(aux)
elif Random_oracle == "constant returning ones, f(i)=1":
    qc.x(aux)
elif Random_oracle == "Balanced function: f(0) = 0 & f(1) = 1":
    qc.cx(i,aux)
elif Random_oracle == "Balanced function: f(1) = 0 & f(0)=1":
    for i in range(len(qr)):
        qc.append(MCXGate(1, ctrl_state='0'), [i, aux]) 

qc.barrier()

reversed_qc = qc.reverse_bits()
psi3 = Statevector.from_instruction(reversed_qc)
display("Psi3", psi3.draw('latex'))

display(qc.draw("mpl"))
```

```python
#Applying Hadamard Gates after the Oracle implementation
for i in range(len(qr)):
    qc.h(i)
    
qc.barrier()

reversed_qc = qc.reverse_bits()
psi4 = Statevector.from_instruction(reversed_qc)
display("Psi4", psi4.draw('latex'))

display(qc.draw("mpl"))
```

```python
#Finally, applying the measurement on Z-basis only of qubits, excluding the auxiliary qubit.

new = qc.measure(range(0,n-1), range(0,n-1))
print(new)
display(qc.draw("mpl"))
```

```python
backend = QasmSimulator()
program_ = qiskit.transpile(qc, backend)
job = backend.run(program_,shots= 2500)
```

```python
result = job.result()
counts = result.get_counts()
print(counts)
```
